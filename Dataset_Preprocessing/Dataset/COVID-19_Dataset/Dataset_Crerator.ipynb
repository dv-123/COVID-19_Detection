```
import pandas as pd
import os
import shutil
```

```
# create the data for postive samples (Github)

FILE_PATH = '/content/drive/My Drive/COVID-19_Project/Dataset_Processing/Covid-19_Chest_X-ray_Dataset/metadata.csv'
IMAGES_PATH = '/content/drive/My Drive/COVID-19_Project/Dataset_Processing/Covid-19_Chest_X-ray_Dataset/images'
```

```
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```
df.head(11)
```

```
TARGET_DIR = "/content/drive/My Drive/COVID-19_Project/Dataset_Processing/Dataset/Covid-19"

if not os.path.exists(TARGET_DIR):
  os.mkdir(TARGET_DIR)
  print("Covid-19 folder created")
```

```
cnt = 0

for (i,row) in df.iterrows():
  if row["finding"] == "Pneumonia/Viral/COVID-19"  and (row['view'] == 'PA' or row['view'] == 'AP'):
    filename = row["filename"]
    image_path = os.path.join(IMAGES_PATH,filename)
    image_copy_path = os.path.join(TARGET_DIR,filename)
    shutil.copy2(image_path, image_copy_path)
    print("Moving image ", cnt)
    cnt = cnt + 1
```

```
print("Total number of covid-19 images collected --> ", cnt)
```

```
import random

KAGGLE_FILE_PATH_NORMAL = "/content/drive/My Drive/COVID-19_Project/Dataset_Processing/Chext_X-ray_Kaggle/train/NORMAL"
KAGGLE_FILE_PATH_PNEUMONIA = "/content/drive/My Drive/COVID-19_Project/Dataset_Processing/Chext_X-ray_Kaggle/train/PNEUMONIA"

TARGET_FILE_PATH_NORMAL = "/content/drive/My Drive/COVID-19_Project/Dataset_Processing/Dataset/Normal"
TARGET_FILE_PATH_PNEUMONIA = "/content/drive/My Drive/COVID-19_Project/Dataset_Processing/Dataset/Pneumonia"
```

since we will be increasing the covid images 2-4 times by using data augmentation resulting 684-1368 images, we will be sampling near to 1368 images from the repective folders also.

sampling images from normal 1340

sampling images from pneumonia 1370

```
image_names_normal = os.listdir(KAGGLE_FILE_PATH_NORMAL)
image_names_pneumonia = os.listdir(KAGGLE_FILE_PATH_PNEUMONIA)
```

```
random.shuffle(image_names_normal)
random.shuffle(image_names_pneumonia)
```

for normal images

```
cnt = 0

for i in range (1340):

  image_name = image_names_normal[i]
  image_path = os.path.join(KAGGLE_FILE_PATH_NORMAL, image_name)

  target_path = os.path.join(TARGET_FILE_PATH_NORMAL, image_name)

  shutil.copy2(image_path, target_path)

  print("Moving image ", cnt)

  cnt = cnt + 1
```

```
cnt = 0

for i in range (1370):

  image_name = image_names_pneumonia[i]
  image_path = os.path.join(KAGGLE_FILE_PATH_PNEUMONIA, image_name)

  target_path = os.path.join(TARGET_FILE_PATH_PNEUMONIA, image_name)

  shutil.copy2(image_path, target_path)

  print("Moving image ", cnt)

  cnt = cnt + 1
```
