AffineTransformation and rotation of the images

```
images_new = os.listdir(SOURCE_PATH_Augmented)
```

```
len(images_new)
```

```
for j in range(len(folders_augmented_all)):
  image_foler_path = os.path.join(SOURCE_PATH_Augmented_all,folders_augmented_all[j])
  images_new = os.listdir(image_foler_path)
  for i in range(len(images_new)):
    image_path = os.path.join(image_foler_path,images_new[i])
    image = imread(image_path)

    # doing only minute changes so that pattern does not change
    # first performing the Affine Transform on the image
    imgTri = np.array( [[0, 0], [image.shape[1] -1, 0], [0, image.shape[0] -1]] ).astype(np.float32)
    dstTri = np.array( [[0, image.shape[1]*0.09], [image.shape[1]*0.97, image.shape[0]*0.07], [image.shape[1]*0.09, image.shape[0]*0.97]] ).astype(np.float32)
    #wrapping the image
    warp_mat = cv2.getAffineTransform(imgTri, dstTri)
    warp_dst = cv2.warpAffine(image, warp_mat, (image.shape[1], image.shape[0]))
    # performing the rotations
    center = (warp_dst.shape[1]//2, warp_dst.shape[0]//2)
    angle = -5
    scale = 1.05
    rot_mat = cv2.getRotationMatrix2D( center, angle, scale )
    warp_rotate_dst = cv2.warpAffine(warp_dst, rot_mat, (warp_dst.shape[1], warp_dst.shape[0]))


    print("Saving image", folders_augmented_all[j] ,i,"_augmented")
    imageio.imwrite(image_foler_path+"/"+'augmented_{}'.format(images_new[i]), warp_rotate_dst)
```

# Performing Image Augmentation on test image

```
image_path = os.path.join(SOURCE_PATH,images[300])
image = imread(image_path)
```

```
plt.imshow(image)
```

```
image.shape
```

```
image.dtype
```

```
src = image
srcTri = np.array( [[0, 0], [src.shape[1] -1, 0], [0, src.shape[0] -1]] ).astype(np.float32)
dstTri = np.array( [[0, src.shape[1]*0.09], [src.shape[1]*0.97, src.shape[0]*0.07], [src.shape[1]*0.09, src.shape[0]*0.97]] ).astype(np.float32)
```

```
warp_mat = cv2.getAffineTransform(srcTri, dstTri)
warp_dst = cv2.warpAffine(src, warp_mat, (src.shape[1], src.shape[0]))
```

```
plt.imshow(warp_dst)
```

```
center = (warp_dst.shape[1]//2, warp_dst.shape[0]//2)
angle = -5
scale = 1.05
rot_mat = cv2.getRotationMatrix2D( center, angle, scale )
warp_rotate_dst = cv2.warpAffine(warp_dst, rot_mat, (warp_dst.shape[1], warp_dst.shape[0]))
```
